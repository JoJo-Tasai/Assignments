### Thinking2:RNN为什么会出现梯度消失？
### answer：因为tanh’ <= 1，训练过程大部分情况下tanh的导数是小于1的,如果Ws也在[0,1]之间，当T很大的时候,W                          在T时刻的梯度就接近于0，即梯度消失。

```python
import torch
from torch import nn, optim
from model import TextRNN
import numpy as np

#数据文件设置
train_file = '/cnews/cnews.train.txt'
test_file = '/cnews/cnews.test.txt'
val_file = '/cnews/cnews.val.txt'
vocab_file = '/cnews/cnews.vocab.txt'

def train():
    model = TextRNN().cuda()
    Loss = nn.MultiLabelSoftMarginLoss()
    optimizer = optim.Adam(model.parameters(),lr=0.001)
    
    for epoch in range(1000):
        print('epoch=', epoch)
        for step,(x_batch, y_batch) in enumerate(train_loader):
            x = x_batch.cuda()
            y = y_batch.cuda()
            out = model(x)
            loss = Loss(out, y)
            print('Loss=', loss)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            accuracy = np.mean((torch.argmax(out, 1) == torch.argmax(y, 1)).cpu().numpy())
            print(accuracy)
```
